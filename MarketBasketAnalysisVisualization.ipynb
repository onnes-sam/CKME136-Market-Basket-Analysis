```python
import pandas as pd
import pandasql as ps
```

```python
Transactions_Results = pd.read_parquet('transactions.parquet', engine='pyarrow')
Transactions_Results['ItemsCount'] = Transactions_Results['items'].apply(len)
```

```python
print(Transactions_Results)
```

```python
TwoOrderedItems = Transactions_Results.loc[Transactions_Results['ItemsCount'] == 2].sort_values(by='freq', ascending=False)
```

```python
from IPython.display import display, HTML
pd.set_option('display.width', 10000)
display(TwoOrderedItems)
```

```python
ThreeOrderedItems = Transactions_Results.loc[Transactions_Results['ItemsCount'] == 3].sort_values(by='freq', ascending=False)
display(ThreeOrderedItems)
```

```python
pd.set_option('display.max_colwidth', -1)
MostOrderedItems = Transactions_Results.sort_values(by='ItemsCount', ascending=False)
display(MostOrderedItems['items'].head(5))
```

```python
DuplicatedItems = Transactions_Results.sort_values(by='ItemsCount', ascending=False)
ids = DuplicatedItems['items']
display(DuplicatedItems[ids.isin(ids[ids.duplicated()])].head(5))
```
